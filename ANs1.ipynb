```python
import numpy as np
import matplotlib.pyplot as plt


data = np.loadtxt('Data_1.txt')


maxima_indices = []
minima_indices = []


for i in range(1, len(data) - 1):
    if data[i] > data[i - 1] and data[i] > data[i + 1]:
        maxima_indices.append(i)
    elif data[i] < data[i - 1] and data[i] < data[i + 1]:
        minima_indices.append(i)


plt.figure(figsize=(10, 6))
plt.plot(data, label='Signal')


plt.plot(maxima_indices, data[maxima_indices], 'ro', label='Maxima')


plt.plot(minima_indices, data[minima_indices], 'bo', label='Minima')


plt.title('Signal with Maxima (red) and Minima (blue)')
plt.xlabel('Index')
plt.ylabel('Signal Value')
plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data (should match the data used in the C program)
data = [1.0, 2.1, 3.0, 1.5, 2.7, 0.5, 2.0, 3.5, 1.0]

# Indices of maxima and minima from C code output
maxima_indices = [2, 7]
minima_indices = [3, 5]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data, label='Data', marker='o')

# Highlight maxima
plt.scatter(maxima_indices, [data[i] for i in maxima_indices], color='red', label='Maxima')

# Highlight minima
plt.scatter(minima_indices, [data[i] for i in minima_indices], color='blue', label='Minima')

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Data with Maxima and Minima')
plt.legend()
plt.grid(True)

# Save the plot to a file
plt.savefig('peaks_plot.png')
plt.show()
```
